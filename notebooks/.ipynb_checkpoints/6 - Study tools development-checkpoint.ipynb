## Hosmer-Lemeshow tests

```python
path_results = 'mimic3models/in_hospital_mortality/predictions/results/TEST.2020-04-14_8-47.k_clstms.none.csv'
results = cb.results_csv(path_results)
results.head(2)
```

Total number expected events

```python
results["prediction"].sum()
```

```python
from scipy.stats import chi2
def hl_ranges(results):
    max_g = int(results["prediction"].sum()*0.9)
    g_list = np.arange(6,max_g)
    pval_list = []
    hltest_list = []
    for g in g_list:
        i, hltest, pval = am.hl_test(results, g, verbose=False)
        pval_list.append(pval)
        hltest_list.append(hltest)
        
    hl_df = pd.Series(hltest_list)
    hl_rolled = hl_df.rolling(11, min_periods=1, center=True, win_type='gaussian').mean(std=5)
    pval_rolled = [1 - chi2.cdf(val, i+1) for i, val in enumerate(hl_rolled)]
    return pval_list, pval_rolled, hltest_list, hl_rolled, g_list
```

## Outcome distributions

absolute distributions of samples with or without event is not possible as the class imbalance is too high --> need to look at proportions instead: group into risk bins of 4% each

```python
def prop_plot(df_):
    df = df_.copy()
    df['rounded risk'] = np.floor(df['prediction']*25)
    prop_event = df.groupby('rounded risk')['y_true'].mean()*100
    plt.figure(figsize=(10,5))
    plt.plot((0,100), (0,100), ls="--", c=".3")
    plt.bar(x = prop_event.index*4+2, height=prop_event.values, width=1)
```

```python
prop_plot(results)
```

```python
prop_plot(results_F)
```

```python
prop_plot(results_M)
```

```python
prop_plot(results_pub)
```

```python
prop_plot(results_priv)
```
