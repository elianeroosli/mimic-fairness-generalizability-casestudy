## 1) Exponential quantiles: visualization of bin size

```python
exp_q = np.linspace(0.0, 1.0, num=10+1)**0.20
diff = [(exp_q[i+1] - exp_q[i])*100 for i in range(10)]

f, ax = plt.subplots(1, 1, figsize=(4, 5))
ax.scatter(x=range(1,11), y=diff, s=30, c='black')
ax.set_xlabel("Group number", fontsize=14)
ax.set_ylabel("Group size (% of total size)", fontsize=14)
ax.tick_params(labelsize=12)
f.savefig('plots/exp_quantiles.jpeg', bbox_inches='tight', transparent=True)
```
